**Conditions**

  In thermodynamic, We can desided if the reaction is spontaneous or Non-spontaneous from the Gibbs free energy change G and calculted by Vant Hoff equation; G = RT ln KD

```python
#-----(Define Variabls)-----#
G = float(input("Enter the Gibbs free energy change G value: "))

#-----(Condition)-----#

if G > 0:
    print("The reaction is Non-spontaneous")
else:
    print("The reaction is Spontaneous")
```

**Loops**

In laboratory chemistry, sometimes we need to prepare several solutions with different concentrations from the stock solution, which are calculated by the dilution law; M1 V1 = M2V2

```python
#------------------------------------(Titl)------------------------------------#
print("Calcultations of preparing chemical solutions using the dilution law ")

#------------------------------(Define Variabls )------------------------------#
M1 = float(input("Enter the concentration of stock solution: "))
V2 = float(input("Enter the volume in mL of final solution: "))

#------------------------------------(Loop)------------------------------------#

for M2 in range(10, 60, 5):  #start from 10 ~60 jumping 10
    V1 = (M2 * V2) / M1
    print(f"To prepare {M2} M of solution in {V2} mL, The required volume of the stock solution = {V1:.2f} mL"  )
```

# Artificial Neural Network

```python
#-----------(Import Libraries)-----------#
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, r2_score

#-----------(1. Data Preparation )-----------#
#-----[1-1. Define features (input) & target (output)]-----#
X = df[['Initial Concentration (mg/L)', 'pH', 'Dossage', 'Contact Time (min)', 'Temperafture C']]
y = df['% Removal Efficiency']

#-----[1-2. Split data into training and testing sets ]-----#
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

#-----[1-3. StandardScaler]-----#
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


#-----------(2. ANN Model Building )-----------#
model = MLPRegressor(hidden_layer_sizes=(64, 32, 16), activation='relu', solver='adam', max_iter=1000, random_state=42)
model.fit(X_train, y_train)

#-----------(3. Model Evaluation )-----------#
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

#-----------(4. Visualization: Actual vs Predicted)-----------#
plt.figure(figsize=(5,3))
plt.scatter(y_test, y_pred, color='violet', label='Predicted', alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='purple', linestyle='--', linewidth=2, label='Perfect Prediction')
plt.title('Actual vs. Predicted Values \n')
plt.xlabel('Actual Removal Efficiency % ')
plt.ylabel('Predicted Removal Efficiency % ')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
#-----------(4-2. Visualization: Actual vs. Predicted for Dosage)-----------#

#-----[4-2-1. Selecting the property for predicted]-----#
x_values = X_test[:, 2]

#-----[4-2-2. Plot for actual & predicted values ]-----#
plt.figure(figsize=(5.5, 4))
plt.scatter(x_values, y_test, color='violet', label='Experimental', s=100, marker='.')
plt.scatter(x_values, y_pred, color='blue', label='ANN Predicted', s=100, marker='.')
plt.title('Removal Efficiency vs. Dosage')
plt.xlabel('Dosage (mg)')
plt.ylabel('Removal Efficiency (%)')
plt.legend()
plt.grid(color='lightgray', linestyle='--', linewidth=0.5, alpha=0.7)
plt.tight_layout()
plt.show()
```
